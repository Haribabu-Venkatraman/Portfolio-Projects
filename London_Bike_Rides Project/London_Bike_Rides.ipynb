```python
# Renaming columns:

col_dict = {
    'timestamp': 'Time',
    'cnt': 'Count',
    't1': 'Temp_real_C',
    't2': 'Temp_feels_C',
    'hum': 'Humidity_percent',
    'wind_speed': 'Wind_speed_Kph',
    'weather_code': 'Weather',
    'is_holiday': 'Holiday',
    'is_weekend': 'Weekend',
    'season': 'Season'
}

df.rename(col_dict, axis=1, inplace=True)
```

```python
df
```

```python
# Changing Humidity value to Percentage:

df.Humidity_percent=df.Humidity_percent/100
```

```python
df
```

```python
# creating weather dictionary to map numeric values to actual written value:

w_dict={
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
} 
```

```python
# Changing datatype before mapping:
df.Weather = df.Weather.astype('str')

# Mapping values:
df.Weather = df.Weather.map(w_dict)
```

```python
df
```

```python
# creating Season dictionary to map numeric values to actual written value:

s_dict={
    '0.0': 'Spring',
    '1.0': 'Summer',
    '2.0': 'Autumn',
    '3.0': 'Winter'
} 
```

```python
# Changing datatype before mapping:
df.Season = df.Season.astype('str')

# Mapping values:
df.Season = df.Season.map(s_dict)
```
